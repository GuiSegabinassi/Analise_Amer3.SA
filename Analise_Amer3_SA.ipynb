```python
df["Adj Close"].value_counts()
```

```python
df["Adj Close"].plot(legend = True, label = "Americanas", figsize = (15, 9))
```

```python
df["MM7"] = df["Adj Close"].rolling(7).mean()
```

```python
display(df)
```

```python
df[["Adj Close", "MM7"]].plot(legend = True, label = "Preço de Fechamendto Ajustado e Média Movel de 7 Dias", figsize = (15, 9))
```

```python
dias = [15, 30, 60]

for dia in dias:
  df["MM" + str(dia)] = df["Adj Close"].rolling(dia).mean()
```

```python
display(df)
```

No Gráfico abaixo temos as médias moveis para 7, 15, 30, 60 dias. Dessa forma podemos observar o preço de fechamento ajustado com menos ruidos

```python
df[["Adj Close", "MM7", "MM15", "MM30", "MM60"]].plot(legend = True, label = "Média Movel", figsize = (15, 9))
```

```python
df["Retorno_diario"] = df["Adj Close"].pct_change()
display(df)
```

```python
df["Retorno_diario"].plot(legend = True, label = "Retorno Diario", figsize = (15, 9))
```

```python
retorno_Amer3 = df.Retorno_diario.dropna().mean()
risco_Amer3 = df.Retorno_diario.dropna().std()
```

```python
display("Retorno_Amer3", retorno_Amer3, "risco_Amer3", risco_Amer3)
```

Aqui temos um Grid que nos mostra o risco e o retorno da Amer3.SA

```python
plt.grid()
plt.xlabel("Média de Retorno Diario = ")
plt.ylabel("Risco Diario")
plt.scatter(retorno_Amer3, risco_Amer3, marker = "*")
```
